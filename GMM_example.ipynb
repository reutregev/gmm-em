```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

from src.data_gen import generate_data
from src.constants import N_CLUSTERS, N_SAMPLES, MAX_ITER
from src.em_algorithm import GMM
```

```python
x, y = generate_data(N_SAMPLES, N_CLUSTERS)
```

```python
gmm = GMM(N_CLUSTERS, MAX_ITER, x)
silhouette_score_gmm = gmm.run(plot_clusters=True, plot_ll=True)
print(f"Silhouette score: {silhouette_score_gmm:.4f}")
```

### Number of clusters by DBSCAN

```python
dbscan = DBSCAN()
dbscan.fit(x)
labels = dbscan.labels_
print(f"Number of clusters by DBSCAN: {sum(np.unique(labels) >= 0)}\nSilhouette score: {silhouette_score(x, labels):.4f}")
```

DBSCAN clustering found that there are 3 core clusters in the data (while 4 clusters were generated).<br>
\* Note that the result of the DBSCAN is of a default setup, i.e., without hyperparameter tuning.
